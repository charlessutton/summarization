# Glove embedding 

In this notebook I show how to create a GLOVE embedding for txt data

 <hr style="border-color:#1e549d"> 

```python
import itertools
from glove import Corpus, Glove
import gensim
```

```python
txt_file = "/home/ubuntu/summarization_query_oriented/data/wikipedia/txt/td_qfs_rank_1_all.txt"
```

```python
with open(txt_file,'r') as f:
    text = f.read()
    
print type(text)
print len(text)
```

```python
sentences = text.decode('utf-8').split('.')
print 'nb of sentences : ' len(sentences)
```

```python
sentences_splitted = []
for sentence in sentences:
    sentences_splitted.append(gensim.utils.simple_preprocess(sentence, deacc=False, min_len=2, max_len=15))
```

```python
sentences_splitted[2]
```

```python
corpus = Corpus()
corpus.fit(sentences_splitted, window=10)
```

```python
glove = Glove(no_components=400, learning_rate=0.05)
glove.fit(corpus.matrix, epochs=30, no_threads=4, verbose=True)
```

```python
glove_dir = "/home/ubuntu/summarization_query_oriented/nn_models/word_vectorization_models/glove/"
glove_name = "rank1_alltxt_size_400_epoch_30.glove"
corpus_name = "rank1_alltxt.corpus"
```

```python
glove.save(glove_dir + glove_name)
corpus.save(glove_dir + corpus_name)
```

```python
glove.add_dictionary(corpus.dictionary)
```

```python
glove.most_similar('iris', number = 10)
```
